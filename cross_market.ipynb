```python
import requests
import pandas as pd
import time

coin_ids = ("bitcoin", "ethereum", "tether")
all_data = []

for coin_id in coin_ids:
    url = f"https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart"
    params = {
        "vs_currency": "usd",
        "days": 365
    }

    response = requests.get(url, params=params)

    #  Check if request worked
    if response.status_code != 200:
        print(f"Error for {coin_id}: {response.status_code}")
        print(response.text)
        continue

    data = response.json()

    #  Check if 'prices' exists
    if "prices" not in data:
        print(f"No price data for {coin_id}")
        print(data)
        continue

    df = pd.DataFrame(data["prices"], columns=["timestamp", "price_usd"])
    df["date"] = pd.to_datetime(df["timestamp"], unit="ms").dt.date
    df["coin_id"] = coin_id
    df = df[["coin_id", "date", "price_usd"]]

    all_data.append(df)

    time.sleep(2)  #  Avoid rate limit

if all_data:
    top3coins_df = pd.concat(all_data, ignore_index=True)
    print(top3coins_df.head())
else:
    print("No data collected.")
```

```python
#5)Get the % change in Bitcoins price between feb 2025 and feb 2026": """
query = """   
SELECT
    (
        MAX(CASE WHEN date = '2026-02-21' THEN price_usd END)
        -
        MAX(CASE WHEN date = '2025-02-22' THEN price_usd END)
    )
    / MAX(CASE WHEN date = '2025-02-22' THEN price_usd END) * 100
    AS percent_change
   FROM top3_coins_price
WHERE coin_id = 'bitcoin'
  AND date IN ('2025-02-22', '2026-02-21');
"""
cursor.execute(query)
cursor.fetchall()
```

```python
#1)Find the highest oil price in the last 5 years.
query = """
SELECT MAX(price) AS highest_price
FROM oil_prices
WHERE `date` BETWEEN DATE_SUB(CURDATE(), INTERVAL 5 YEAR) AND CURDATE();
"""
cursor.execute(query)
cursor.fetchall()
```

```python
#2)Get the average oil price per year.
query = """
SELECT 
    YEAR(`date`) AS year,
    ROUND(AVG(price), 2) AS avg_price
FROM oil_prices
GROUP BY YEAR(`date`)
ORDER BY year;
"""

cursor.execute(query)
cursor.fetchall()
```

```python
#3)Show oil prices during COVID crash (MarchApril 2020).
query = """
SELECT `date`, price
FROM oil_prices
WHERE `date` BETWEEN '2020-03-01' AND '2020-04-30'
ORDER BY `date`;
"""

cursor.execute(query)
cursor.fetchall()
```

```python
#4)Find the lowest price of oil in the last 10 years.
query = """
SELECT 
    MIN(price) AS lowest_price
FROM oil_prices
WHERE `date` >= DATE_SUB(CURDATE(), INTERVAL 10 YEAR);
"""

cursor.execute(query)
cursor.fetchone()
```

```python
#5)Calculate the volatility of oil prices (max-min difference per year).
query = """
SELECT 
    YEAR(`date`) AS year,
    ROUND(MAX(price) - MIN(price), 2) AS volatility
FROM oil_prices
GROUP BY YEAR(`date`)
ORDER BY year;
"""

cursor.execute(query)
cursor.fetchall()
```

```python
###Join Queries (Cross-Market Analysis)
```

```python
#1)Compare Bitcoin vs Oil average price in 2025.
query = """
WITH btc_avg_2025 AS (
    SELECT
        AVG(price_usd) AS avg_btc_2025
    FROM top3_coins_price 
    WHERE 
        coin_id = 'bitcoin'
        AND date BETWEEN '2025-02-18' AND '2025-12-31'
),
oil_avg_2025 AS (
    SELECT
        AVG(Price) AS avg_oil_2025
    FROM oil_prices
    WHERE
        Date BETWEEN '2025-01-01' AND '2025-12-31'
)

SELECT
    btc_avg_2025.avg_btc_2025,
    oil_avg_2025.avg_oil_2025
FROM btc_avg_2025
CROSS JOIN oil_avg_2025;
"""
df = pd.read_sql(query, engine)
print(df)
```

```python
#2)Check if Bitcoin moves with S&P 500 (correlation idea).
query = """
SELECT
    b.date,
    b.price_usd AS bitcoin_price,
    s.Close AS sp500_price
FROM top3_coins_price b
JOIN stock_indices s
    ON b.date = s.Date
WHERE b.coin_id = 'bitcoin'
  AND s.ticker = '^GSPC'
ORDER BY b.date;
"""

df = pd.read_sql(query, engine)

print("Rows returned:", df.shape[0])
df.head()
```
